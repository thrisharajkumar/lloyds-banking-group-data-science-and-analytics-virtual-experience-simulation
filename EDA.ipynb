```python
import pandas as pd

# Load Excel file
file_path = "Customer_Churn_Data_Large.xlsx"
dfs = pd.read_excel(file_path, sheet_name=None)  # load all sheets

# Extract each sheet
demo = dfs["Customer_Demographics"]
trans = dfs["Transaction_History"]
service = dfs["Customer_Service"]
online = dfs["Online_Activity"]
churn = dfs["Churn_Status"]

# -------------------------------
# STEP 1: Aggregate transaction data
trans_summary = trans.groupby("CustomerID").agg(
    TotalSpend=("AmountSpent", "sum"),
    AvgSpend=("AmountSpent", "mean"),
    NumTransactions=("TransactionID", "count"),
    LastTransaction=("TransactionDate", "max"),
    UniqueProducts=("ProductCategory", "nunique")
).reset_index()

# -------------------------------
# STEP 2: Aggregate customer service data
service_summary = service.groupby("CustomerID").agg(
    NumInteractions=("InteractionID", "count"),
    LastInteraction=("InteractionDate", "max"),
    IssuesUnresolved=("ResolutionStatus", lambda x: (x == "Unresolved").sum())
).reset_index()

# -------------------------------
# STEP 3: Merge all datasets on CustomerID
merged = (demo
          .merge(trans_summary, on="CustomerID", how="left")
          .merge(service_summary, on="CustomerID", how="left")
          .merge(online, on="CustomerID", how="left")
          .merge(churn, on="CustomerID", how="left"))

print("Final merged dataset shape:", merged.shape)
print(merged.head())

# -------------------------------
# STEP 4: Save to CSV
merged.to_csv("merged.csv", index=False)
print("Saved merged dataset as merged.csv")
```

```python
import pandas as pd

# Load merged dataset
df = pd.read_csv("merged.csv")

# 1) Create a flag for customers with no recorded service history
df["ServiceMissing"] = df["NumInteractions"].isna().astype(int)

# 2) Impute service-related fields (domain-driven)
#    - If there was no service history, treat as zeros / "NoInteraction"
df["NumInteractions"]  = df["NumInteractions"].fillna(0)
df["IssuesUnresolved"] = df["IssuesUnresolved"].fillna(0)
df["LastInteraction"]  = df["LastInteraction"].fillna("NoInteraction")

# 3) Sanity checks
print("Nulls after imputation:")
print(df[["NumInteractions","IssuesUnresolved","LastInteraction","ServiceMissing"]].isnull().sum())

print("\nServiceMissing distribution (0=had history, 1=none):")
print(df["ServiceMissing"].value_counts(dropna=False))

print("\nChurn rates by ServiceMissing flag:")
print(
    df.groupby("ServiceMissing")["ChurnStatus"]
      .agg(['count','mean'])
      .rename(columns={'count':'n_customers','mean':'churn_rate'})
)

# Optional: quick spot-check of a few rows
print("\nSample rows:")
print(df[["CustomerID","NumInteractions","IssuesUnresolved","LastInteraction","ServiceMissing","ChurnStatus"]].head(10))

# 4) Save cleaned file
df.to_csv("merged_cleaned.csv", index=False)
print("\nSaved cleaned dataset as merged_cleaned.csv")
```

```python
import pandas as pd
from datetime import datetime

# Load cleaned dataset
df = pd.read_csv("merged_cleaned.csv")

# -----------------------------
# 1. Feature Engineering - Ratios
df["SpendPerTransaction"] = df["TotalSpend"] / df["NumTransactions"].replace(0, 1)
df["SpendPerProduct"] = df["TotalSpend"] / df["UniqueProducts"].replace(0, 1)
df["InteractionRate"] = df["NumInteractions"] / df["NumTransactions"].replace(0, 1)

# -----------------------------
# 2. Feature Engineering - Recency
# Use reference date as 2023-12-31
ref_date = pd.to_datetime("2023-12-31")

# Convert LastTransaction, LastLoginDate, LastInteraction to datetime (if not already)
for col in ["LastTransaction", "LastLoginDate", "LastInteraction"]:
    df[col] = pd.to_datetime(df[col], errors="coerce")

# Calculate recency features
df["DaysSinceLastTransaction"] = (ref_date - df["LastTransaction"]).dt.days
df["DaysSinceLastLogin"] = (ref_date - df["LastLoginDate"]).dt.days
df["DaysSinceLastInteraction"] = (ref_date - df["LastInteraction"]).dt.days

# For customers with "NoInteraction", fill DaysSinceLastInteraction with a large number
df["DaysSinceLastInteraction"].fillna(df["DaysSinceLastInteraction"].max() + 30, inplace=True)

# -----------------------------
# 3. Save updated dataset
df.to_csv("merged_featured.csv", index=False)
print("Feature-engineered dataset saved as merged_featured.csv")
print("New columns added:", [
    "SpendPerTransaction", "SpendPerProduct", "InteractionRate",
    "DaysSinceLastTransaction", "DaysSinceLastLogin", "DaysSinceLastInteraction"
])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# 1) Load data
df = pd.read_csv("merged_cleaned.csv")

# Make sure target is numeric (helps downstream)
df["ChurnStatus"] = pd.to_numeric(df["ChurnStatus"], errors="coerce").fillna(0).astype(int)

# -------------------------------
# 2) Engineer features
df["SpendPerTransaction"] = df["TotalSpend"] / df["NumTransactions"].replace(0, 1)
df["SpendPerProduct"]     = df["TotalSpend"] / df["UniqueProducts"].replace(0, 1)
df["InteractionRate"]     = df["NumInteractions"] / (df["NumTransactions"] + 1)

# Parse dates
for col in ["LastTransaction", "LastLoginDate", "LastInteraction"]:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce")

# Reference date = latest timestamp we have
reference_date = df[["LastTransaction", "LastLoginDate", "LastInteraction"]].max().max()

df["DaysSinceLastTransaction"] = (reference_date - df["LastTransaction"]).dt.days
df["DaysSinceLastLogin"]       = (reference_date - df["LastLoginDate"]).dt.days
df["DaysSinceLastInteraction"] = (reference_date - df["LastInteraction"]).dt.days

# If someone has no interaction timestamp, keep large recency so model can learn it
if df["DaysSinceLastInteraction"].isna().any():
    big_val = df["DaysSinceLastInteraction"].max(skipna=True) + 30
    df["DaysSinceLastInteraction"] = df["DaysSinceLastInteraction"].fillna(big_val)

new_features = [
    "SpendPerTransaction", "SpendPerProduct", "InteractionRate",
    "DaysSinceLastTransaction", "DaysSinceLastLogin", "DaysSinceLastInteraction"
]

# -------------------------------
# 3) Histograms
sns.set_theme(style="whitegrid")
plt.figure(figsize=(16, 10))
for i, col in enumerate(new_features, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[col], bins=30, kde=True, color="royalblue")
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()

# -------------------------------
# 4) Boxplots by churn (BLUE=0, RED=1)
# Key fix: use order=[0,1] and palette list in that order (no dict)
plt.figure(figsize=(16, 10))
for i, col in enumerate(new_features, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(
    data=df,
    x="ChurnStatus", 
    y=col,
    hue="ChurnStatus",          # <- fixes warning
    dodge=False,                # prevent duplicate boxes
    legend=False, 
    palette={0: "blue", 1: "red"}
)

    plt.title(f"{col} by ChurnStatus")
    plt.xlabel("ChurnStatus (0=Active=Blue, 1=Churned=Red)")
plt.tight_layout()
plt.show()



# -------------------------------
# 5) Correlation heatmap
plt.figure(figsize=(12, 8))
corr = df[new_features + ["ChurnStatus"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0, fmt=".2f")
plt.title("Correlation Heatmap (Engineered Features + Churn)")
plt.show()
```

## Categorical churn analysis.

```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_features = ["Gender", "MaritalStatus", "IncomeLevel", "ServiceUsage"]

# -------------------------------
# 1. Countplots for churn by category
plt.figure(figsize=(16, 12))
for i, col in enumerate(categorical_features, 1):
    plt.subplot(2, 2, i)
    sns.countplot(x=col, hue="ChurnStatus", data=df, palette={0: "blue", 1: "red"})
    plt.title(f"Churn Distribution by {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.legend(title="ChurnStatus", labels=["Active (0)", "Churned (1)"])
plt.tight_layout()
plt.show()

# -------------------------------
# 2. Percentage stacked bar plots
for col in categorical_features:
    churn_rate = pd.crosstab(df[col], df["ChurnStatus"], normalize="index") * 100
    churn_rate.plot(kind="bar", stacked=True, figsize=(7, 5), color=["blue", "red"])
    plt.title(f"Churn % by {col}")
    plt.ylabel("Percentage")
    plt.xlabel(col)
    plt.legend(title="ChurnStatus", labels=["Active (0)", "Churned (1)"])
    plt.show()
```
